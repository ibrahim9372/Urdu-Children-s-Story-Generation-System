```python
import re
from collections import Counter

with open("C:\\Users\\hassa\\OneDrive\\Desktop\\sem6\\NLP\\Assignement 1\\urdu_tokenizer_training.txt", 'r', encoding="utf-8") as f:
    text = f.read()

text = re.sub(r'\b(?!(?:EOS|EOP|EOT)\b)[a-zA-Z]+\b', '', text)
raw_words = text.split()

corpus = Counter()

for word in raw_words:
    if word in ['<EOS>', '<EOP>', '<EOT>']:
        char_tuple = (word,)
    else:
        char_tuple = tuple(list(word)) + ('_',)
    
    corpus[char_tuple] += 1
print(dict(list(corpus.items())[:10]))
unique_chars = set()

for word_tuple in corpus.keys():
    for char in word_tuple:
        unique_chars.add(char)

vocab_list = list(unique_chars)

vocab_list.sort()

print(f"Total Unique Characters: {len(vocab_list)}")
print(vocab_list)
```

```python
def get_stats(corpus):
    pairs = {}
    for word_tuple, freq in corpus.items():
        for i in range(len(word_tuple) - 1):
            pair = (word_tuple[i], word_tuple[i+1])
            pairs[pair] = pairs.get(pair, 0) + freq
    return pairs

pairs = get_stats(corpus)
best_pair = max(pairs, key=pairs.get)
best_pair
```

```python
def merge_corpus(pair, corpus, new_id):
    new_corpus = {}
    v1, v2 = pair
    for word_tuple, freq in corpus.items():
        new_word_tuple = []
        i = 0
        while i < len(word_tuple):
            if i < len(word_tuple) - 1 and word_tuple[i] == v1 and word_tuple[i+1] == v2:
                new_word_tuple.append(new_id) 
                i += 2
            else:
                new_word_tuple.append(word_tuple[i])
                i += 1
        new_corpus[tuple(new_word_tuple)] = freq
    return new_corpus
```

```python
token_id={j:i for i,j in enumerate(vocab_list)}
id_token={i:j for i,j in enumerate(vocab_list)}
merge_table={}
id_corpus = {}

for word_tuple, freq in corpus.items():
    id_tuple = tuple(token_id[char] for char in word_tuple)
    id_corpus[id_tuple] = freq

next_id = len(vocab_list) 
for i in range(300):
    all_pair = get_stats(id_corpus)
    if not all_pair: break
    
    pair = max(all_pair, key=all_pair.get)

    merge_table[pair] = next_id
    
    new_string = id_token[pair[0]] + id_token[pair[1]]
    id_token[next_id] = new_string
    token_id[new_string] = next_id
    id_corpus = merge_corpus(pair, id_corpus, next_id)    
    next_id += 1
```
